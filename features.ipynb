```python
import matplotlib.pyplot as plt
import pandas as pd

# Your feature importances
features = ['sqft_living', 'grade', 'lat', 'sqft_living15', 'bathrooms', 
            'zipcode', 'long', 'sqft_lot', 'sqft_lot15', 'view']

importance = [0.1987, 0.1948, 0.1857, 0.1171, 0.0574, 
              0.0415, 0.0398, 0.0323, 0.0322, 0.0321]

# Create dataframe and sort
df = pd.DataFrame({'Feature': features, 'Importance': importance})
df = df.sort_values('Importance', ascending=True)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.barh(df['Feature'], df['Importance'], color='steelblue')

# Highlight top 3
colors = ["#b12cb8" if x >= 0.15 else "#dba6db" for x in df['Importance']]
ax.barh(df['Feature'], df['Importance'], color=colors)

# Add percentage labels
for i, v in enumerate(df['Importance']):
    ax.text(v + 0.005, i, f'{v*100:.1f}%', va='center', fontsize=10)

ax.set_xlabel('Importance (%)')
ax.set_title('What Drives Price: High Value Homes', fontsize=14, fontweight='bold')
ax.set_xlim(0, 0.35)

plt.tight_layout()
plt.savefig('high_value.png', dpi=150, bbox_inches='tight')
plt.show()
```
